```python
from pandas_ml import ConfusionMatrix
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
File=np.load('resnet50_pred2.npz')
#File=np.load('resnet50_CwoD.npz')
Prediction1=File['name1'] 
Prediction2=File['name2']
Prediction3=File['name3']
y_train1=File['name4']
y_valid1=File['name5']
y_test1=File['name6']
classes=File['name7']
```

```python
def find(array,ele):return np.where(array==ele)
```

```python
classes[26]
```

```python
out_classes=classes[Prediction1]
Predictions=y_train1
cm1 = ConfusionMatrix(Predictions,out_classes)
ConfusionMatrix1=cm1.stats()['cm']
ClassStatistics1=cm1.stats()['class'].transpose()
OverallStatistics1=cm1.stats()['overall']
```

```python
OverallStatistics1
```

```python
cm1.plot(normalized=True)
```
